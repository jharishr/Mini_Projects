```python
import zipfile
import os

# Get uploaded file name
zip_filename = next(iter(uploaded))

# Create a folder to extract
extracted_folder = "rgb_images"
os.makedirs(extracted_folder, exist_ok=True)

# Extract
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

print("✅ Zip extracted to:", extracted_folder)
```

```python
from PIL import Image
import os

input_folder = "rgb_images/FILTERED_BRAIN_CT"
output_folder = "grayscale_images"
os.makedirs(output_folder, exist_ok=True)

# Convert to 8-bit grayscale and resize to 64x64
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".png", ".jpg", ".jpeg", ".bmp")):
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)

        # Convert to grayscale + resize to 64x64
        img = img.convert("L").resize((64, 64), Image.BILINEAR)

        img.save(os.path.join(output_folder, filename))

print("✅ All images converted to 8-bit grayscale and resized to 64×64.")
```
